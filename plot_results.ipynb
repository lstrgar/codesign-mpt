```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
pretraining_loss_fpath = ""
zeroshot_loss_fpath = ""
zeroshot_diversity_fpath = ""
fewshot_loss_fpath = ""
fewshot_diversity_fpath = ""
baseline_loss_fpath = ""
baseline_diversity_fpath = ""

pretraining_loss = np.load(pretraining_loss_fpath).mean(-1)
zeroshot_loss = np.load(zeroshot_loss_fpath).mean((1, 2))
fewshot_loss = np.load(fewshot_loss_fpath).mean((1, 2))
baseline_loss = np.load(baseline_loss_fpath).mean((1, 2))
losses = [baseline_loss, pretraining_loss, zeroshot_loss, fewshot_loss]
loss_labels = ["Baseline", "Pretraining", "Zero-Shot", "Few-Shot"]

zeroshot_diversity = np.load(zeroshot_diversity_fpath)
fewshot_diversity = np.load(fewshot_diversity_fpath)
baseline_diversity = np.load(baseline_diversity_fpath)
diversity = [baseline_diversity, zeroshot_diversity, fewshot_diversity]
diversity_labels = ["Baseline", "Zero-Shot", "Few-Shot"]
```

```python
fig1, ax1 = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
for i, (d, l) in enumerate(zip(diversity, diversity_labels)):
    ax1[i].plot(d, color='C2')
    ax1[i].set_xlabel("Generation")
    if i == 0:
        ax1[i].set_ylabel("Diversity")
    ax1[i].set_title(f"{l} Diversity")
plt.show()


fig2, ax2 = plt.subplots(1, 4, figsize=(14, 3), sharey=True)
for i, (l, label) in enumerate(zip(losses, loss_labels)):
    ax2[i].plot(l, color='C1')
    ax2[i].set_xlabel("Generation")
    if i == 0:
        ax2[i].set_ylabel("Test Loss")
    ax2[i].set_title(f"{label} Loss")
plt.show()
```
